```python
sales = [
    {"fruit": "apple",  "qty": 12, "price": 30},
    {"fruit": "orange", "qty": 7,  "price": 20},
    {"fruit": "banana", "qty": 20, "price": 10},
    {"fruit": "apple",  "qty": 5,  "price": 30},
    {"fruit": "mango",  "qty": 3,  "price": 40},
]

stock_left = {"apple": 4, "orange": 15, "banana": 2, "mango": 9}
reorder_level = 5
```

```python
sold = {}

for sale in sales:
    fruit = sale["fruit"]
    qty = sale["qty"]

    if fruit in sold:
        sold[fruit] += qty
    else:
        sold[fruit] = qty

print(sold)
```

```python
best_seller = max(sold, key=sold.get)
print("Best seller:", best_seller)
```

```python
reorder = []

for fruit, stock in stock_left.items():
    if stock <= reorder_level:
        reorder.append(fruit)

print("Reorder now:", reorder)
```

```python
total_revenue = 0

for sale in sales:
    total_revenue = total_revenue + sale["qty"] * sale["price"]

print("Total revenue:", total_revenue)
```

```python
def day_summary():
    print("Total revenue: ₹", total_revenue)
    print("Sold per fruit:", sold)
    print("Best seller:", best_seller)
    print("Reorder now:", reorder)

day_summary()
```

```python
import numpy as np

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

prices = np.array([30, 20, 10, 40])

q = np.array([[12,  5,  8, 10, 14,  9,  6],
              [ 7, 10,  6,  8,  5, 11,  4],
              [20, 15, 18, 22, 25, 19, 12],
              [ 3,  4,  2,  5,  6,  3,  1]])

daily_totals = q.sum(axis=0)
```

```python
bestday = np.argmax(daily_totals)
print(days[bestday])
```

```python
percentages = (daily_totals / daily_totals.sum() * 100).round(1)
```

```python
print(percentages)
```
