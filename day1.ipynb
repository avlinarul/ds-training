```python
prices = {
    "apple": 50,
    "banana": 20,
    "orange": 30
}

basket = ["apple", "banana", "orange"]

total = 0

for item in basket:
    total += prices[item]

print("Basket total:", total)
```

```python
def final_price(price, percent):
    discount = price * percent / 100
    return price - discount

print(final_price(1000, 10))
print(final_price(500, 20))
```
