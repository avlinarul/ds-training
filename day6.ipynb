```python
#8
import pandas as pd

# Load data
sales = pd.read_csv("sales_week.csv")
fruits = pd.read_csv("fruits.csv")

# Clean fruit names
sales["fruit"] = sales["fruit"].str.strip().str.lower()
fruits["fruit"] = fruits["fruit"].str.strip().str.lower()

# Revenue
sales["revenue"] = sales["qty"] * sales["price"]

# Date conversion
sales["date"] = pd.to_datetime(sales["date"])

# Weekday column
sales["weekday"] = sales["date"].dt.day_name()

# Best and worst weekday
weekday_revenue = sales.groupby("weekday")["revenue"].sum()
print("Best weekday:", weekday_revenue.idxmax())
print("Worst weekday:", weekday_revenue.idxmin())

# Daily revenue
daily_revenue = (
    sales.set_index("date")["revenue"]
         .resample("D")
         .sum()
)

print("\nDaily Revenue")
print(daily_revenue)

# Join fruit information
merged = sales.merge(fruits, on="fruit", how="left")

# Handle missing cost price (e.g. lemon)
merged["cost_price"] = merged["cost_price"].fillna(
    merged["cost_price"].mean()
)

# Profit calculation
merged["profit"] = (
    merged["revenue"]
    - (merged["qty"] * merged["cost_price"])
)

# Profit per supplier
profit_supplier = merged.groupby("supplier")["profit"].sum()
print("\nProfit per Supplier")
print(profit_supplier)

# Profit per category
profit_category = merged.groupby("category")["profit"].sum()
print("\nProfit per Category")
print(profit_category)

# Revenue pivot table
pivot = pd.pivot_table(
    merged,
    values="revenue",
    index="weekday",
    columns="stall",
    aggfunc="sum",
    fill_value=0
)

print("\nRevenue Pivot Table")
print(pivot)

# Clean check
print("\nFruit names tidy:",
      merged["fruit"].equals(
          merged["fruit"].str.strip().str.lower()
      ))
```
