```python
import pandas as pd

# Load data
df = pd.read_csv("sales_week.csv")

# Clean column names
df.columns = df.columns.str.strip().str.lower()

# Create revenue column if it doesn't exist
if "revenue" not in df.columns:
    df["revenue"] = df["qty"] * df["price"]

print(df.head())
print(df.columns)
# Total revenue
print("Total Revenue:", df["revenue"].sum())

# Revenue per stall
print(df.groupby("stall")["revenue"].sum())
```

```python
sales = df.groupby(["fruit", "stall"])["revenue"].sum().unstack(fill_value=0)

sales["south_pct"] = (
    sales["South"] /
    sales.sum(axis=1) * 100
).round(1)

print(sales["south_pct"])
```

```python
rev = (
    df.groupby(["stall", "fruit"])["revenue"]
      .sum()
      .reset_index()
)

rev["rank"] = (
    rev.groupby("stall")["revenue"]
       .rank(method="dense", ascending=False)
)

print(rev.sort_values(["stall", "rank"]))
```

```python
fruit_day = (
    df.groupby(["fruit", "date"])["revenue"]
      .sum()
      .reset_index()
)
```

```python
best_day = fruit_day.loc[
    fruit_day.groupby("fruit")["revenue"].idxmax()
]

print(best_day)
```
