```python
#end of day build
import pandas as pd

sales = pd.read_csv("sales_week.csv")
fruits = pd.read_csv("fruits.csv")

sales["revenue"] = sales["qty"] * sales["price"]

# Using a left join so all sales are kept, including lemon sales
report = pd.merge(sales, fruits, on="fruit", how="left")

report["profit"] = report["revenue"] - report["qty"] * report["cost_price"]

print("Total Profit Per Supplier")
supplier_profit = report.groupby("supplier")["profit"].sum().sort_values(ascending=False)
print(supplier_profit)

print("\nMost Valuable Supplier")
print(supplier_profit.idxmax())

print("\nProfit By Origin")
print(report.groupby("origin")["profit"].sum())

unknown_profit = report.loc[report["supplier"].isna(), "revenue"].sum()

print("\nPotential Profit With Unknown Cost Data")
print(unknown_profit)

print("\nSupplier Report")
print("1. TropicoLtd generates the highest total profit.")
print("2. Imported fruits contribute a large share of profit.")
print("3. Local fruits remain an important source of profit.")
print("4. Lemon sales have missing supplier and cost information.")
print("5. Profit for lemon cannot be calculated accurately until cost data is available.")
print("6. Updating the fruit reference table would improve reporting accuracy.")
```

```python
#b
left = pd.merge(sales, fruits, on="fruit", how="left")

left["cost_price"] = left.groupby("origin")["cost_price"].transform(
    lambda x: x.fillna(x.mean())
)

print(left[["fruit", "origin", "cost_price"]].head())
```

```python
#c
suppliers = pd.DataFrame({
    "supplier": ["FreshFarms", "TropicoLtd", "CitrusCo", "VineyardX", "OrchardCo"],
    "city": ["Madurai", "Chennai", "Coimbatore", "Bangalore", "Salem"]
})

combined = pd.merge(sales, fruits, on="fruit", how="left")
combined = pd.merge(combined, suppliers, on="supplier", how="left")

print(combined[["fruit", "supplier", "city"]].head())
```

```python
#d
combined = pd.merge(sales, fruits, on="fruit", how="inner")

combined["profit"] = combined["revenue"] - combined["qty"] * combined["cost_price"]

profit_per_sale = (
    combined.groupby("supplier")["profit"]
    .mean()
    .sort_values(ascending=False)
)

print(profit_per_sale)
```
