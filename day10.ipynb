```python
import pandas as pd
import sqlite3
 
sales = pd.read_csv("sales_week.csv")
sales["revenue"] = sales["qty"] * sales["price"]
 
conn = sqlite3.connect("stall.db")            # a database file
sales.to_sql("sales", conn, if_exists="replace", index=False)
 
```

```python
def run(sql):
    return pd.read_sql(sql, conn)
 
run("SELECT * FROM sales LIMIT 5")
```

```python
# Q2: All sales at the North stall over ₹300 revenue, biggest first
run("""
SELECT *,
       qty * price AS revenue
FROM sales
WHERE stall = 'North'
  AND qty * price > 300
ORDER BY revenue DESC
""")
```

```python
# Q3: The 5 biggest single sales of the week
run("""
SELECT *,
       qty * price AS revenue
FROM sales
ORDER BY revenue DESC
LIMIT 5
""")
```

```python
# Q4: All grape and mango sales priced ₹40 or more
run("""
SELECT *
FROM sales
WHERE fruit IN ('grape', 'mango')
  AND price >= 40
""")
```

```python
# Q5: All sales of fruits whose name contains "an"
run("""
SELECT *
FROM sales
WHERE fruit LIKE '%an%'
""")
```

```python
# Q6: All sales with qty between 10 and 20, sorted by revenue
run("""
SELECT *,
       qty * price AS revenue
FROM sales
WHERE qty BETWEEN 10 AND 20
ORDER BY revenue DESC
""")
```

```python
# Q7: Every fruit with a price of ₹30 or more, listed once each
run("""
SELECT DISTINCT fruit
FROM sales
WHERE price >= 30
""")
```

```python
# Stretch (a): Label each sale as premium or regular
run("""
SELECT fruit,
       price,
       CASE
           WHEN price >= 40 THEN 'premium'
           ELSE 'regular'
       END AS tier
FROM sales
""")
```

```python
# Stretch (b): Fruits with exactly 5 letters
run("""
SELECT *
FROM sales
WHERE LENGTH(fruit) = 5
""")
```

```python
# Stretch (c): Premium fruits at North or any grape
run("""
SELECT *
FROM sales
WHERE (price >= 40 AND stall = 'North')
   OR fruit = 'grape'
""")
```
