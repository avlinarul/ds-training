```python
import pandas as pd

# Load the file
df = pd.read_csv("sales.csv")

# Confirm rows and columns
print("Shape:", df.shape)

# Add revenue column
df["revenue"] = df["qty"] * df["price"]

# Total revenue
print("\nTotal Revenue:")
print(df["revenue"].sum())

# Sales count by fruit
print("\nSales Count by Fruit:")
print(df["fruit"].value_counts())

# Biggest sale
print("\nBiggest Sale:")
print(df.sort_values("revenue", ascending=False).head(1))

# Fruits priced ₹40 or more
print("\nSales with Price >= ₹40:")
print(df[df["price"] >= 40])

# Week in Review
print("\nWeek in Review")
print("----------------")
print(f"Rows and Columns: {df.shape}")
print(f"Total Revenue: ₹{df['revenue'].sum()}")
print(f"Average Price: ₹{df['price'].mean():.2f}")
print(f"Most Common Fruit: {df['fruit'].value_counts().idxmax()}")
print(f"Highest Revenue Sale: ₹{df['revenue'].max()}")
```
