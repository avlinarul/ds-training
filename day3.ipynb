```python
import pandas as pd
import numpy as np

df = pd.read_csv("sales.csv")

# Rows and columns
print("Shape:", df.shape)

# Revenue column
df["revenue"] = df["qty"] * df["price"]
print("Total Revenue:", df["revenue"].sum())

# Sales count by fruit
print("\nSales Per Fruit:")
print(df["fruit"].value_counts())

# Biggest sale
print("\nBiggest Sale:")
print(df.nlargest(1, "revenue"))

# Top 3 sales
print("\nTop 3 Sales:")
print(df.nlargest(3, "revenue"))

# Tier column
df["tier"] = np.where(df["price"] >= 40, "premium", "regular")

# Premium vs Regular revenue
premium_revenue = df[df["tier"] == "premium"]["revenue"].sum()
regular_revenue = df[df["tier"] == "regular"]["revenue"].sum()

print("\nPremium Revenue:", premium_revenue)
print("Regular Revenue:", regular_revenue)

# Apple sales with quantity above 10
print("\nApple Sales (Qty > 10):")
print(df[(df["fruit"] == "apple") & (df["qty"] > 10)])

# Week in review
print("\nWeek in Review")
print("1. Total weekly revenue:", df["revenue"].sum())
print("2. Most sold fruit:")
print(df["fruit"].value_counts().head(1))
print("3. Highest revenue sale identified.")
print("4. Top 3 sales ranked by revenue.")
print("5. Premium and regular revenue compared.")
print("6. Apple sales above quantity 10 filtered.")
```
