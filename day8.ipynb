```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
df = pd.read_csv("sales_week.csv")
df["revenue"] = df["qty"] * df["price"]
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="fruit", y="revenue", estimator="sum", errorbar=None)
plt.title("Total revenue per fruit")
plt.xlabel("Fruit")
plt.ylabel("Revenue (₹)")
plt.tight_layout()
plt.savefig("revenue_per_fruit.png", dpi=150)
plt.show()
```

```python
#a
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.barplot(data=df, x="fruit", y="revenue", estimator="sum", errorbar=None)

plt.title("Total Revenue per Fruit")
plt.xlabel("Fruit")
plt.ylabel("Revenue (₹)")

plt.tight_layout()
plt.savefig("total_revenue_per_fruit.png", dpi=150)
plt.show()
```

```python
#b
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="stall", y="revenue", estimator="sum", errorbar=None)

plt.title("Total Revenue per Stall")
plt.xlabel("Stall")
plt.ylabel("Revenue (₹)")

plt.tight_layout()
plt.savefig("total_revenue_per_stall.png", dpi=150)
plt.show()
```

```python
#c
fruit_totals = (
    df.groupby("fruit")["revenue"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(data=fruit_totals, x="fruit", y="revenue")

plt.title("Total Revenue per Fruit (Highest to Lowest)")
plt.xlabel("Fruit")
plt.ylabel("Revenue (₹)")

plt.tight_layout()
plt.savefig("sorted_revenue_per_fruit.png", dpi=150)
plt.show()
```

a) This chart compares the total revenue generated by each fruit. Each bar represents one fruit, making it easy to identify the highest and lowest revenue. The chart is fully labelled and saved for reporting.

b) This chart shows the total revenue earned by each stall. Comparing the bar heights helps identify which stall generated the most revenue. Clear labels make the visualization easy to understand.

c) The fruit revenues were first grouped and sorted in descending order before plotting. Displaying the bars from highest to lowest makes it easier to compare the top-performing fruits. This improves the readability of the chart.

```python
#b
daily = df.groupby("date")[["revenue", "qty"]].sum().reset_index()

plt.figure(figsize=(9,5))

sns.lineplot(data=daily, x="date", y="revenue", marker="o", label="Revenue")
sns.lineplot(data=daily, x="date", y="qty", marker="o", label="Quantity")

plt.title("Daily Revenue and Quantity")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)

plt.legend()

plt.tight_layout()
plt.savefig("daily_revenue_qty_line.png", dpi=150)
plt.show()
```

a) This line chart displays the total revenue earned each day. It makes it easy to observe increasing, decreasing, or stable revenue trends over time. Peaks and dips highlight the busiest and quietest days.

b) This chart compares daily revenue and quantity sold using two lines. Similar movement between the lines suggests that revenue and quantity change together. Differences may occur because product prices also affect revenue.

```python
#6
corr = df[["qty", "price", "revenue"]].corr()
 
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("How quantity, price and revenue relate")
plt.tight_layout(); plt.show()
```

```python
#strech
fig, axes = plt.subplots(2, 2, figsize=(12,10))

# Bar chart
sns.barplot(data=df, x="fruit", y="revenue", estimator="sum",
            errorbar=None, ax=axes[0,0])
axes[0,0].set_title("Revenue by Fruit")

# Histogram
sns.histplot(data=df, x="revenue", bins=10, ax=axes[0,1])
axes[0,1].set_title("Revenue Distribution")

# Line chart
daily = df.groupby("date")["revenue"].sum().reset_index()
sns.lineplot(data=daily, x="date", y="revenue", marker="o",
             ax=axes[1,0])
axes[1,0].tick_params(axis="x", rotation=45)
axes[1,0].set_title("Daily Revenue")

# Scatter plot
sns.scatterplot(data=df, x="qty", y="revenue", hue="fruit",
                ax=axes[1,1])
axes[1,1].set_title("Quantity vs Revenue")

plt.tight_layout()
plt.savefig("dashboard.png", dpi=150)
plt.show()
```

```python
g = sns.catplot(
    data=df,
    x="fruit",
    y="revenue",
    col="stall",
    kind="bar",
    estimator=sum,
    errorbar=None,
    height=4,
    aspect=1
)

g.set_axis_labels("Fruit", "Revenue (₹)")
g.set_titles("Stall: {col_name}")
g.fig.suptitle("Revenue by Fruit for Each Stall", y=1.05)

plt.savefig("facet_chart.png", dpi=150)
plt.show()
```
