```python
def bill_with_tax(amount):
    tax = amount * 0.05
    return amount + tax

print(bill_with_tax(1000))
print(bill_with_tax(250))

def make_receipt(customer, amount):
    return f"{customer} paid ₹{amount}. Thank you!"

print(make_receipt("Ravi", 560))
```
