The main questions explored include which fruit, stall, category, weekday, and supplier performed best during the week.

3. Data Cleaning & Preparation

```python
import pandas as pd
import matplotlib.pyplot as plt

sales = pd.read_csv("sales_week.csv")
fruits = pd.read_csv("fruits.csv")
```

```python
sales.info()
sales.isnull().sum()
sales.duplicated().sum()
```

```python
df = pd.merge(sales, fruits, on="fruit")
```

```python
df["Revenue"] = df["qty"] * df["price"]

df["Profit"] = (df["price"] - df["cost_price"]) * df["qty"]
```

Cleaning Summary

• Loaded both datasets.

• Joined them using the fruit column.

• Checked missing values.

• Checked duplicate records.

• Created Revenue and Profit columns.

Exploratory Analysis

Analysis 1

Revenue by Fruit

Question:

Which fruit generated the highest revenue?

```python
fruit = df.groupby("fruit")["Revenue"].sum()

fruit.plot(kind="bar")

plt.title("Revenue by Fruit")

plt.xlabel("Fruit")

plt.ylabel("Revenue")

plt.show()
```

Grape generated the highest revenue during the week, making it the best-performing fruit.

Analysis 2

Revenue by Stall

```python
stall = df.groupby("stall")["Revenue"].sum()

stall.plot(kind="bar")

plt.title("Revenue by Stall")

plt.xlabel("Stall")

plt.ylabel("Revenue")

plt.show()
```

Takeaway

South stall generated higher revenue than North stall during the week.

Analysis 3

Revenue by Weekday

```python
weekday = df.groupby("date")["Revenue"].sum()

weekday.plot(kind="line", marker="o")

plt.title("Revenue by Day")

plt.xlabel("Day")

plt.ylabel("Revenue")

plt.show()
```

Takeaway

Thursday recorded the highest revenue during the week.

Analysis 4

Revenue by Category

```python
category = df.groupby("category")["Revenue"].sum()

category.plot(kind="bar")

plt.title("Revenue by Category")

plt.xlabel("Category")

plt.ylabel("Revenue")

plt.show()
```

Takeaway

The Citrus category generated the highest revenue during the week.

Analysis 5

Profit by Supplier

```python
supplier = df.groupby("supplier")["Profit"].sum()

supplier.plot(kind="bar")

plt.title("Profit by Supplier")

plt.xlabel("Supplier")

plt.ylabel("Profit")

plt.show()
```

Takeaway

The supplier with the highest total profit contributed the most to overall business profitability.
