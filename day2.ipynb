```python
fruits = ["apple", "orange", "banana", "mango"]
prices = np.array([30, 20, 10, 40])
 
# rows = fruits, columns = Mon..Sun (7 days)
q = np.array([[12,  5,  8, 10, 14,  9,  6],   # apple
              [ 7, 10,  6,  8,  5, 11,  4],   # orange
              [20, 15, 18, 22, 25, 19, 12],   # banana
              [ 3,  4,  2,  5,  6,  3,  1]])  # mango
```

```python
print(q.sum(axis=1))
```

```python
qty=(q.sum(axis=1))
print(qty)
```

```python
totalrevenue=qty*prices
print(totalrevenue)
```

```python
total=(totalrevenue.sum())
print(total)
```

```python
qtyperday=(q.sum(axis=0))
print(qtyperday)
```

```python
print(qtyperday>40)
```

```python
daysabove40=qtyperday[qtyperday>40]
print(daysabove40)
```

```python
count=(qtyperday>40).sum()
print(count)
```
